# DATA CLEANING

```python
import pandas as pd
import numpy as np
data=pd.read_excel('emlakjetIstanbulIlceleri.xlsx')
df=data.copy()
```

```python
df.head()  #first 5 rows in df
```

```python
df.shape # rows and columns
```

```python
# Drop columns which are useless
df = df.drop(columns=['Unnamed: 0'])
df=df.drop(columns=[0,1,3,5,6,7,9,10,13,14,15])
```

```python
df.isnull().values.any()  #ıs there any null values?
```

```python
df.isnull().values.sum() 
```

```python
df.isnull().sum() #total of  null values 
```

```python
df = df.dropna() #drop null values from df

# Reindex index number
df[''] = np.arange(len(df))
df = df.set_index('')
```

```python
df.shape
```

```python
df.columns=['type', 'room','size','price','location'] #rename the column names
df.head(10)
```

```python
home_type = ['Daire', 'home', 'Residence'] #  get only Daire,home and Residence 
df = df[df['type'].isin(home_type)]
```

```python
#df[df['room']=='Stüdyo']
df.loc[df['room'] == 'Stüdyo', 'room'] = '1+0' # convert the Stüdyo houses to 1+0
```

```python
room_type=['1+0','1+1','2+1','3+1','4+1','5+1','5+2']
df = df[df['room'].isin(room_type)]
```

```python
# split to city district and Neigborhood
location_split = df['location'].str.split(' - ', expand=True)
location_split.columns = ['City', 'District', 'Neighborhood']
df = pd.concat([df, location_split], axis=1)
```

```python
df= df.drop(columns=['location'])
df.head()
```

```python
# Reindex index number
df[''] = np.arange(len(df))
df = df.set_index('')
```

```python
df['size'] = df['size'].str.replace('m2', '')  
df['price'] = df['price'].str.replace('TL', '')
df['price'] = df['price'].str.replace('.', '')
```

```python
df.head()
```

```python
# Convert multiple object columns to integers
df['price'] = df['price'].astype(int)
df['size'] = df['size'].astype(int)
```

```python
city=['İstanbul']
df = df[df['City'].isin(city)]
```

```python
df=df[df['size']>30] # drppped rows that less than 30 size and less than 2000 price and bigger than 200.000
df=df[df['size']<1000]
df=df[df['price']>2000]
df=df[df['price']<220000]
df=df[df['City']=='İstanbul']
# Reindex index number
df[''] = np.arange(len(df))
df = df.set_index('')

#### Filtering Rent,Deposit and Dues Values. ###
#df= df.loc[(df["Rent"] >= 5000) & (df["Rent"] <= 80000),:]
```

# DATA VISUALIZATION

```python
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
# average price for each home type
y = df.groupby("room")["price"].mean()
x = np.array(['1+0','1+1','2+1','3+1','4+1','5+1','5+2'])

plt.grid(axis="y")
plt.title("average price for each home type")
plt.bar(x,y)
plt.show()
```

```python
# Assuming df already contains the 'room' and 'size' columns
# Calculate the average size for each room type
average_size_by_room = df.groupby("room")["size"].mean()

# Define the room types in the order you want them to appear
room_types = ['1+0', '1+1', '2+1', '3+1', '4+1', '5+1', '5+2']

# Ensure the x-values match the room types in the correct order
x = np.array(room_types)
y = np.array([average_size_by_room.get(room, 0) for room in room_types])

# Plotting the scatter plot
plt.figure(figsize=(8, 4))
plt.scatter(x, y, s=100, c='blue', alpha=0.6)
plt.grid(True)
plt.title("Average m² for Each Home Type", fontsize=16, weight="bold")
plt.xlabel("Room Type", fontsize=14)
plt.ylabel("Average Size (m²)", fontsize=14)

# Annotate the points with their values
for i, txt in enumerate(y):
    plt.annotate(f'{txt:.1f}', (x[i], y[i]), textcoords="offset points", xytext=(0,10), ha='center')

plt.show()

```

```python
unique_room_list = df["room"].unique()
unique_room_list
```

```python
# count of home type
# the code below can also write using 'groupby' method!
unique_room_list = df["room"].unique()

twoone= [] # mean of "twoone" is "2+1" house
fourone= []
treeone= []
oneone= []
onezero= []
fiveone=[]
fivetwo=[]

for i in range(len(df)):
    if unique_room_list[0] == df["room"][i]:
        twoone.append(df["room"][i])
    elif unique_room_list[1] == df["room"][i]:
        treeone.append(df["room"][i])
    elif unique_room_list[2] == df["room"][i]:
        oneone.append(df["room"][i])
    elif unique_room_list[3] == df["room"][i]:
        fourone.append(df["room"][i])
    elif unique_room_list[4] == df["room"][i]:
        onezero.append(df["room"][i])
    elif unique_room_list[5] == df["room"][i]:
        fiveone.append(df["room"][i])
    else:
        fivetwo.append(df["room"][i])

print(f"count of 4+1 house: {len(fourone)}")
print(f"count of 3+1 house: {len(treeone)}")
print(f"count of 2+1 house: {len(twoone)}")
print(f"count of 1+1 house: {len(oneone)}")
print(f"count of 1+0 house: {len(onezero)}")
print(f"count of 5+1 house: {len(fiveone)}")
print(f"count of 5+2 house: {len(fivetwo)}")
```

```python
# count of home type
# determination x and y labels of pie chart 
x = np.array(["5+2","5+1","4+1", "3+1", "2+1", "1+1", "1+0"])
y = np.array([len(fivetwo), len(fiveone), len(fourone), len(treeone), len(twoone), len(oneone), len(onezero)])

# Define blue tones
blue_tones = ['#2384F3', '#4882C4', '#8896A6', '#c5e0ed', '#295383', '#B4C1D0', '#75a0cf']

# Create pie chart with blue tones
plt.figure(figsize=(5, 5))
plt.pie(y, labels=x, autopct='%1.1f%%', colors=blue_tones)
plt.title('Distribution of Istanbul Rental Flat Rooms', weight="bold")

plt.show()
```

```python
# Average of rental house prices for each cities

average_prices = df.groupby("District")["price"].mean()
y = average_prices.values
x = average_prices.index

plt.figure(figsize=(8, 10))
plt.grid(linewidth=0.4,  axis="x")

plt.barh(x, y,  height=0.5,  color=["blue","orange"])

plt.title("average rent price of each cities")

plt.show()
```

```python
# Average of rental house prices for each cities
import matplotlib.pyplot as plt

# Calculate the average prices by district
average_prices = df.groupby("District")["price"].mean()

# Sort the average prices in descending order and select the top 5
top_5_districts = average_prices.sort_values(ascending=False).head(7)

# Extract values and indices for plotting
y = top_5_districts.values
x = top_5_districts.index

# Plotting
plt.figure(figsize=(6, 4))  # Adjusted figure size to fit the top 5 districts
plt.grid(linewidth=0.4, axis="x")
plt.barh(x, y, height=0.5, color=["orange", "blue"])
plt.title("Top 5 Districts with Highest Average Rent Price")
plt.xlabel("Average Rent Price")
plt.ylabel("District")

plt.show()
```

```python
# Average of rental house prices for each cities
import matplotlib.pyplot as plt

# Calculate the average prices by district
average_prices = df.groupby("District")["price"].mean()

# Sort the average prices in descending order and select the top 5
top_5_districts = average_prices.sort_values(ascending=False).tail()

# Extract values and indices for plotting
y = top_5_districts.values
x = top_5_districts.index

# Plotting
plt.figure(figsize=(6, 4))  # Adjusted figure size to fit the top 5 districts
plt.grid(linewidth=0.4, axis="x")
plt.barh(x, y, height=0.5, color=["orange", "blue"])
plt.title("Top 5 Districts with Lowest Average Rent Price")
plt.xlabel("Average Rent Price")
plt.ylabel("District")

plt.show()
```
